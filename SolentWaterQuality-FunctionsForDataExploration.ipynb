## wqf.DAIN: plots 111/117, dividing by seasons, locations etc.

Dissolved available inorganic nitrogen (DAIN) is calculated as the sum of ammonia (N) and total
oxidized N. Total oxidized N is itself the sum of nitrate and nitrite. 

Orthophosphate results were used as a direct measure of dissolved available inorganic phosphorus (DAIP). Two sets of determinands for several of these values were extracted from WIMS as the data recording changed over time.

Nutrient loads were calculated as nutrient concentration multiplied by flow.
